```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x=np.arange(1,6)
x
```

```python
y=np.arange(2,11,2)
y
```

```python
#plot(): 
plt.plot(x,y,"red")
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,"blue")
plt.subplot(2,2,2)
plt.plot(y,x,"red")
plt.subplot(2,2,3)
plt.plot(y,y**2,"green")
plt.subplot(2,2,4)
plt.plot(x,x**2,"yellow")
plt.show()
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.2,0.3])

axes1.plot(y,x)
axes1.set_title("Outer Graph")

axes2.plot(x,y)
axes2.set_title("Inner Graph")

```

```python
fig,axes=plt.subplots(nrows=2,ncols=1)
plt.tight_layout()
```

```python
#axes ler üzerinde gezinme
fig,axes=plt.subplots(nrows=2,ncols=1)
axes[0].plot(x+2,y)
axes[0].set_title("First Axes")
axes[1].plot(x,y**3)
axes[1].set_title("Second Axes")
plt.tight_layout()
```

```python
#axes ler üzerinde gezinme
fig,axes=plt.subplots(nrows=2,ncols=1)
for ax in axes:
    ax.plot(x+2,y**3)
plt.tight_layout()
```

```python
fig=plt.figure(figsize=(5,5))
plt.plot(x,y,color="red")
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(7,5))
axes[0].plot(x**5,y,color="purple")
axes[0].set_title("First Axes")
axes[1].plot(x,y**3,color="green")
axes[1].set_title("Second Axes")
plt.tight_layout()
```

```python
#figure kaydetme
fig.savefig("Figure1.png")
```

```python
fig=plt.figure(figsize=(6,4))
axes=fig.add_axes([0,0,1,1]) #hepsini kapsayan bir figure oluşturur
axes.plot(x,x**0.5,color="blue",label="x karakök")    #x in karakökünün grafiği
axes.plot(x,x**2,color="green",label="x kare")       #x in karesinin grafiği
axes.plot(x,x**3,color="purple",label="x küp")       #x in küpünün grafiği
axes.legend() #etiketleri ekranda göstermek için 
```

```python
fig=plt.figure(figsize=(6,6))
axes=fig.add_axes([0,0,1,1]) #hepsini kapsayan bir figure oluşturur
axes.plot(x,x**2,color="blue",label="x kare",linewidth = 4) 
axes.legend()
```

```python
fig=plt.figure(figsize=(6,6))
axes=fig.add_axes([0,0,1,1]) #hepsini kapsayan bir figure oluşturur
axes.plot(x,x**2,color="blue",label="x kare",linewidth = 4,linestyle = "--") 
axes.legend()
```

```python
fig=plt.figure(figsize=(6,6))
axes=fig.add_axes([0,0,1,1]) #hepsini kapsayan bir figure oluşturur
axes.plot(x,x**2,color="blue",label="x kare",linewidth = 4,linestyle = ":") 
axes.legend()
```

```python
fig=plt.figure(figsize=(6,4))
axes=fig.add_axes([0,0,1,1]) #hepsini kapsayan bir figure oluşturur
axes.plot(x,x**0.5,color="blue",label="x karakök",linewidth = 4,linestyle = ":")    #x in karakökünün grafiği
axes.plot(x,x**2,color="green",label="x kare",linewidth = 6,linestyle = "-.")       #x in karesinin grafiği
axes.plot(x,x**3,color="purple",label="x küp",linewidth = 3,linestyle = "--")       #x in küpünün grafiği
axes.legend()
fig.savefig("Fig1.png")
```
