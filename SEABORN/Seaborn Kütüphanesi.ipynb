# heatmap()

```python
#türkiyenin yıllara göre enflasyon bilgisi
tufe=pd.read_csv("tüfe.csv")
tufe
```

```python
enf=tufe.pivot_table(index="ay",columns="yıl",values="enflasyon")
enf
```

```python
sns.heatmap(enf)
#sıcaklık haritası oluşturur
```

```python
sns.heatmap(enf,annot=True)
```

```python
sns.heatmap(enf,annot=True,linecolor="black",lw="1")
```

# clustermap()

```python
#benzerliklere göre sınıflandırma(kümeleme) yapar
sns.clustermap(enf,annot=True,linecolor="black",lw="1")
```

```python
sns.clustermap(enf,figsize=(10,7))
```

# PairGrid()

```python
dunya=pd.read_csv("dünyam.csv")
dunya=dunya.dropna(how="any")
dunya
```

```python
dunya=sns.PairGrid(dunya)
dunya.map_diag(sns.distplot)
dunya.map_upper(plt.scatter)
dunya.map_lower(sns.kdeplot)
```
